### What is a Pandas DataFrame?

A **Pandas DataFrame** is a two-dimensional, size-mutable, and heterogeneous tabular data structure with labeled axes (rows and columns). It's similar to an Excel spreadsheet or a SQL table. Each column in a DataFrame is essentially a **Pandas Series**. In essence, a DataFrame is like a collection of Pandas Series objects that share the same index, representing the rows of the DataFrame.

---

### Why Do We Need Pandas DataFrame?

While a **Pandas Series** is useful for handling one-dimensional data (similar to a single column in a table), real-world data often has more complexity and is organized in rows and columns. For instance, data might have multiple attributes such as names, ages, and cities, which need to be represented in a tabular format with multiple columns. 

A **DataFrame** allows you to:
1. **Store multiple Series together**: A DataFrame is like a collection of Series, where each Series represents a column, and the DataFrame allows you to work with the entire dataset in a structured way.
2. **Label rows and columns**: Just like a Series has an index, a DataFrame has both row indices and column labels, enabling easy access and manipulation of data based on names instead of positions.
3. **Perform operations on multiple columns**: A DataFrame allows operations that involve multiple Series (columns), such as filtering, aggregating, and merging data across columns.

---

### Key Features of Pandas DataFrame:

1. **Tabular Structure**: 
   - Each column is a Series, and all columns share the same row indices.
   - It allows for easy data organization and manipulation in a table-like structure, making it perfect for tasks like data cleaning and analysis.

2. **Heterogeneous Data**: 
   - Each column can have different data types (integers, floats, strings, etc.), unlike a Series where all elements are of the same type. This flexibility is crucial for handling real-world datasets where different attributes of data (e.g., age, income, name) are of different types.

3. **Labeled Index and Columns**: 
   - Rows and columns have labels (indices), allowing intuitive data selection and operations. You can access data by the row index or column name, making DataFrame operations more descriptive and user-friendly.

4. **Handling Missing Data**: 
   - DataFrames offer robust support for handling missing values, with tools to detect, fill, or drop missing data. This is especially important in real-world datasets where missing or incomplete information is common.

---

### How Are Pandas Series and DataFrames Related?

- **A DataFrame is a collection of Series**: Each column in a DataFrame is essentially a Series. So, a DataFrame can be thought of as a container for multiple Series that share the same row index.
  
- **Dimensionality**:
   - **Series**: One-dimensional (only rows).
   - **DataFrame**: Two-dimensional (rows and columns).

- **Index Alignment**: Series and DataFrames can be combined or compared because they are both labeled structures. For example, you can add a Series to a DataFrame (or between two DataFrames), and Pandas will align the indices (both row and column labels) before performing the operation.

- **Example Use Case**: Suppose you have data on employee names and their salaries. A Series could be used to represent employee names or salaries, but to analyze both together, you would use a DataFrame, with one Series representing names and another Series representing salaries.

---

### When to Use a Pandas Series vs. a Pandas DataFrame?

- **Use Pandas Series** when:
  - You need to work with a single column of data (like a list of names or a list of prices).
  - You need to perform simple operations on a one-dimensional array-like structure.
  
- **Use Pandas DataFrame** when:
  - You need to work with multiple columns of data, and relationships between different attributes matter.
  - You are dealing with tabular data (multiple attributes for each record).
  - You need to perform operations involving multiple Series at once, such as merging, aggregating, or filtering by conditions across columns.

---

### Summary:

- **Pandas Series**: A one-dimensional labeled array, similar to a column in a spreadsheet.
- **Pandas DataFrame**: A two-dimensional labeled data structure, similar to a table or a spreadsheet, where each column is a Series, and the columns can have different data types and also each row can also be seen as a pandas series
  
- **Relation**: A DataFrame is essentially a collection of Pandas Series that are aligned by row indices. Series are the building blocks of DataFrames, providing a foundation for creating more complex, multidimensional data structures in Pandas.

```python
# Write a funtion that can return track record of two teams against each other
ipl['WinningTeam'].unique() # so no tie match is there
```

```python
def track_record(team1, team2):
    # Masks to find matches between team1 and team2 in both possible configurations
    mask1 = (ipl['Team1'] == team1) & (ipl['Team2'] == team2)
    mask2 = (ipl['Team2'] == team1) & (ipl['Team1'] == team2)

    # Combine both masks
    mask = mask1 | mask2

    # Count the number of wins for team1 and team2 in those matches
    count1 = ipl[mask & (ipl['WinningTeam'] == team1)].shape[0]
    count2 = ipl[mask & (ipl['WinningTeam'] == team2)].shape[0]

    # Return the win counts for both teams
    return [count1, count2]
```

```python
print(track_record('Chennai Super Kings', 'Mumbai Indians'))
```

```python
points_df = pd. DataFrame(
    {
        '1st point': [(3,4), (-6,5), (0,0), (-10,1), (4,5)],
        '2nd point': [(-3,4), (0,0), (2,2), (10,10), (1,1)]
    }
)
points_df
```

```python
def dist(row):
    pt_a = row['1st point']
    pt_b = row['2nd point']
    return ( (pt_a[0] - pt_b[0])**2 + (pt_a[1] - pt_b[1])**2 )**0.5
```

```python
points_df.apply(dist, axis=1) # VV IMP AXIS=1 SE YE COLUMNS PE APPLY NHI JO BY DEFAULT PAR ROW PE APPLY
```

```python
points_df['distance'] = points_df.apply(dist, axis=1)
points_df
```
