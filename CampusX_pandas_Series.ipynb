### What is Pandas?

Pandas is a fast, powerful, and flexible open-source data analysis and manipulation library in Python. It is designed to work with structured data and provides easy-to-use data structures for handling large datasets. Pandas is widely used for tasks like data cleaning, exploration, and transformation.

### Why Use Pandas?
- **Data manipulation:** Efficiently handles structured data with tools for reshaping, merging, and slicing.
- **Data analysis:** Provides powerful functionalities for summarizing, grouping, and analyzing datasets.
- **Easy handling of missing data:** Functions to detect, fill, and drop missing values.
- **Integration with Python:** Built on top of Python, making it easy to integrate with other libraries like NumPy and Matplotlib.
  
---

### Pandas Data Structures

#### 1. **Pandas Series:**
   - **Definition:** A one-dimensional labeled array, similar to a column in a spreadsheet or SQL table. It can hold data of any type (integer, float, string, etc.).
   - **Indexing:** Each element has a label (index) for easy access.
   - **Example:** 
     ```python
     import pandas as pd
     s = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
     print(s)
     ```

#### 2. **Pandas DataFrame:**
   - **Definition:** A two-dimensional, size-mutable, and heterogeneous tabular data structure with labeled axes (rows and columns), similar to an Excel spreadsheet or SQL table.
   - **Why Use:** Allows handling large datasets, applying operations over rows/columns, and data selection/filtering.
   - **Example:** 
     ```python
     df = pd.DataFrame({
         'Name': ['Alice', 'Bob', 'Charlie'],
         'Age': [25, 30, 35],
         'City': ['New York', 'Los Angeles', 'Chicago']
     })
     print(df)
     ```

The key difference between `sort_values()` and `sort_index()` in a Pandas Series is **what is being sorted**:

### 1. **`sort_values()`**:
   - **Purpose:** Sorts the Series based on the values (data) in ascending or descending order.
   - **Usage:**
     ```python
     series.sort_values(ascending=True, inplace=False)
     ```
   - **Example:**
     ```python
     import pandas as pd
     s = pd.Series([3, 1, 2], index=['b', 'c', 'a'])
     s_sorted = s.sort_values()
     print(s_sorted)
     ```
     **Output:**
     ```
     c    1
     a    2
     b    3
     dtype: int64
     ```

### 2. **`sort_index()`**:
   - **Purpose:** Sorts the Series based on its index (labels) in ascending or descending order.
   - **Usage:**
     ```python
     series.sort_index(ascending=True, inplace=False)
     ```
   - **Example:**
     ```python
     s_sorted_index = s.sort_index()
     print(s_sorted_index)
     ```
     **Output:**
     ```
     a    2
     b    3
     c    1
     dtype: int64
     ```

### Summary:
- **`sort_values()`** orders the Series by the **data**. here based on movie name column
- **`sort_index()`** orders the Series by the **index/labels**. here baed on movie lead name

````python
# Find actors who have scored more than 20 movies
movies[movies.value_counts() > 20]
````

````python
num_movies = movies.value_counts()
num_movies
````

````python
num_movies[num_movies > 20]
````

The error `IndexingError: Unalignable boolean Series provided as indexer` occurs because the index of the boolean Series (`movies.value_counts() > 20`) doesn't match the index of the original `movies` Series.

### Why the Error Occurs:
In the first case (`movies[movies.value_counts() > 20]`), the boolean condition (`movies.value_counts() > 20`) creates a new Series with a different index, which doesn’t align with the index of the original `movies` Series. This mismatch causes the error.

### Why the Second Case Works:
In the second case (`num_movies[num_movies > 20]`), you're working directly with the result of `value_counts()`. Since the indexing happens within the same object (`num_movies`), there’s no issue of misaligned indices, so the operation works correctly.
